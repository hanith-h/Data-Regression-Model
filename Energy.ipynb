'''x for independent columns y for target variable '''

```python
# Remove the 'Cooling Load' column from df1 and store the rest in x
x=df1.drop(columns=['Cooling Load'])
```

```python
#Viewing columns in x
x.columns
```

```python
# Store the 'Cooling Load' column from df1 in y (Target)
y=df1['Cooling Load']
```

```python
y #Calling y
```

```python
#Importing SKLEARN
from sklearn.feature_selection  import SelectKBest,f_regression
```

```python
# Select all features using the f_regression scoring function
sk=SelectKBest(score_func=f_regression,k='all')
```

```python
# Select the best features from x based on y
x_new=sk.fit_transform(x,y)
```

```python
# Get the names of the selected features
selected_columns=x.columns[sk.get_support()]
```

```python
#Calling selected_columns
selected_columns
```

```python
x_selected=x[selected_columns]
```

```python
# Create a DataFrame with feature names and their scores, sorted by score
feature_score=pd.DataFrame({"Feature":x.columns,'Scores':sk.scores_}).sort_values(
    by='Scores',ascending=False)
```

```python
#Calling Feature Score
feature_score
```

Scaling
-------

```python
#Importing Standard Scaler
from sklearn.preprocessing import StandardScaler
```

```python
#Setting up the Scaler
ss=StandardScaler()
```

```python
#Scaling the selected features
x_scaled=ss.fit_transform(x_selected)
```

```python
#Viewing the scaled ones
x_scaled
```

```python
#Importing train_test_split
from sklearn.model_selection import train_test_split
```

```python
# Split the data into training and testing sets
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.2,random_state=42)
```

```python
# Print the shapes of the train and test sets
print(' x_train shape:', x_train.shape, '\n', 'x_test shape:', x_test.shape, '\n', 'y_train shape:', y_train.shape, '\n', 'y_test shape:', y_test.shape)
```

```python
#Importing the ALGORITHM--------------------------*Linear Regression*
from sklearn.linear_model import LinearRegression
```

```python
# Create a LinearRegression object
lr=LinearRegression()
```

```python
#Trainning data
lr.fit(x_train,y_train)
```

```python
#Predicting
ypred_lr=lr.predict(x_test)
```

```python
#Importing mean_absolute_error,mean_squared_error,r2_score
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
#Evaluating predicted values (mean_absolute_error,mean_squared_error,r2_score)
mae=mean_absolute_error(y_test,ypred_lr) 
msc=mean_squared_error(y_test,ypred_lr)
r2s=r2_score(y_test,ypred_lr)
```

```python
# Displaying the predicted values
print('mae:',mae,'\nmsc:',msc,'\nr2s',r2s)
```

```python
#Importing the ALGORITHM--------------------------*SVR*
from sklearn.svm import SVR
```

```python
# Create a SVRRegression object
svr=SVR()
```

```python
#Training
svr.fit(x_train,y_train)
```

```python
#Predicting
ypred_svr=svr.predict(x_test)
```

```python
#Evaluating predicted values (mean_absolute_error,mean_squared_error,r2_score)
mae_svr=mean_absolute_error(y_test,ypred_svr) 
msc_svr=mean_squared_error(y_test,ypred_svr)
r2s_svr=r2_score(y_test,ypred_svr)
```

```python
# Displaying the predicted values
print('mae:',mae_svr,'\nmsc:',msc_svr,'\nr2s',r2s_svr)
```

```python
from sklearn.tree import DecisionTreeRegressor
```
